## Importing libraries

```python
import numpy as np
import pandas as pd
```

```python
credit = pd.read_csv('tmdb_5000_credits.csv')
movie = pd.read_csv('tmdb_5000_movies.csv')
credit.head()
```

```python
movie.head()
```

```python
## Printing the movie and credit shape
print("Credits:", credit.shape)
print("Movies:",movie.shape)
```

```python
credit_renamed_column = credit.rename(index=str, columns={"movie_id":"id"})
movie_merge = movie.merge(credit_renamed_column,on='id')
print(movie_merge.head())
```

```python
movie_cleaned_df = movie_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
print(movie_cleaned_df.head())
print(movie_cleaned_df.info())
print(movie_cleaned_df.head(1)['overview'])
```

## Content based recommendation system

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

### Fitting the tf-idf on the overview-text

```python
tfv_matrix = tfv.fit_transform(movie_cleaned_df['overview'].values.astype('U'))
print(tfv_matrix)
print(tfv_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

### Computing the sigmoid model

```python
sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
print(sig[0])
```

### Reverse mapping of indices and movie titles

```python
indices = pd.Series(movie_cleaned.index, index=movie_cleaned['original_title']).drop_duplicates()
print(indices)
print(indices['The Dark Knight Rises'])
print(sig[4799])
print(list(enumerate(sig[indices['The Dark Knight Rises']])))
print(sorted(list(enumerate(sig[indices['The Dark Knight Rises']])), key=lambda x: x[1], reverse=True))
```

```python
def give_recommendation(title, sig=sig):
    # Get corresponding title index
    idx = indices[title]
    # Get pairwise similarity scores
    sig_scores = list(enumerate(sig[idx]))
    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    # Scores of the top 7 similar movies
    sig_scores = sig_scores[1:8]
    # Movie Indices
    movie_indices = [i[0] for i in sig_scores]
    # Top similar movies
    return movie_cleaned['original_title'].iloc[movie_indices]
    
```

```python
inp = input("Enter the movie you've watched:")
results = print("Similar movies that you should watch:", give_recommendation(inp))
```
